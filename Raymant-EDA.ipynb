```python
# extracting required features for our data analysis from the given dataset
```

```python
df_n=df[['A','B','C','G','R','X6','X8','Class']]
```

```python
print(df['Class'].value_counts())
sns.countplot(df['Class'])
```

 there are 6 classes [1-6]

 there is imbalance between the classes , we will perform data analysis separately for individual classes. 

```python
for i,e in enumerate(df_n.drop("Class",axis=1).columns):
    plt.figure(i)
    sns.boxplot(df[e],df_n['Class'])
    plt.show()
```

### using boxplot-
    class A has high impact of features B,C,X6,X8.
    class B has high impact of features G,X8.
    class C has high impact of features B,C,R,X6,X8.
    class D has high impact of features A,D.
    class E has high impact of features A,B,C,G. 
    class F has high impact of features B,C,R,X6.

```python
# creating new dataset for each class of our target column
```

```python
df1=df_n[df_n['Class']=='A']
df2=df_n[df_n['Class']=='B']
df3=df_n[df_n['Class']=='C']
df4=df_n[df_n['Class']=='D']
df5=df_n[df_n['Class']=='E']
df6=df_n[df_n['Class']=='F']
```
